```python
#В файле представлены финансы:Порядок:Числовой;Банкрот:Y/N;Студент:Y/N;Баланс:Числовой;Заработок:Числовой; Total Rows:1000	Total Columns:5
import pandas as pd #Импорт модуля pandas как переменную pd
import numpy as np #Импорт модуля numpy как переменную np
import matplotlib.pyplot as plt #Импорт модуля matplotlib.pyplot как переменную plt
import seaborn as sns #Импорт модуля seaborn как переменную sns
import matplotlib as mpl #Импорт модуля matplotlib как переменную mpl
file = '/content/Default.xlsx' #Создание переменной для загрузки файла в типе файла excel (elsx) 
df = pd.read_excel(file) #Создание переменной и прочиткой с помощью pandas загруженного файла в типе файла excel (xlsx)
df.replace({'Yes': 1, 'No': 0}, inplace=True) #Используем модуль replace для соответсвия Yes - 1,No - 0 для дальнейшего анализа без NaN в описании
df.tail(10) # Изображение статистики по данному файлу (head-начало,tail-хвост), значение в '()' - кол-во. вывод строк
```

```python
headers=['A', 'Банкрот', 'Студент', 'Баланс','Заработк'] #Создает заголовки по имеющимся 5-тю колонкам в excel файле Default.xlsx 
df.columns = headers #Сводим колонны с заголовками
df.drop('A', axis=1, inplace=True)
df.head(4) #Выводим 5 первых строчек
```
