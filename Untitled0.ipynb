```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
```

```python
data_flight = sns.load_dataset("mpg") # загрузка датасета mpg
data_flight.head() # изображение даты
sns.scatterplot(data=data_flight, x="model_year",y="horsepower") #линейный график с "годом" на оси x и "лошадиными силами" на оси y.
plt.show() # изображение даты
sns.lineplot(data=data_flight, x="model_year", y="mpg") #диаграмма рассеяния с "годом" на оси x и "потреблением топлива" на оси y.
plt.show() # изображение даты
sns.barplot(data=data_flight, x="model_year", y="cylinders") # столбчатая диаграмма с "годом" на оси x и "цилиндрами" на оси y.
plt.show() # изображение даты
sns.distplot(data_flight["model_year"]) # гистограмма столбца "год"
plt.show() # изображение даты
```

```python
sns.set_style("dark") #изменение на темную тему
diamonds_data = sns.load_dataset('diamonds')  #загрузка датыбазы
plt.subplot(1, 2, 1)  # 1 строка, 2 столбца 2 графика (размеры графика)
sns.countplot(x='carat', data=diamonds_data) #караты
plt.subplot(1, 2, 2)
sns.countplot(x='depth', data=diamonds_data) #глубина
plt.show()
sns.heatmap(diamonds_data .corr(), annot=True) #annot=True позволяет выводить численные значения в каждом квадрате, представляющем значение корреляции(стат.связь) для corr матрицы
plt.show()
```
